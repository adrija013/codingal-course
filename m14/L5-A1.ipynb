```python
import seaborn as sns
import pandas as pd
```

```python
weather = pd.read_csv('Weather Dataset.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
weather.isnull().sum()
```

```python
weather.dropna(axis=1, how='all', inplace=True)
```

```python
weather.isnull().sum()
```

```python
sns.barplot(data=weather, x="weather_type", y="temperature")
```

```python
sns.histplot(data=weather, x='temperature', kde=True)
```

```python
sns.distplot(weather['humidity'], rug=True)
```

```python
sns.jointplot(data=weather, x="humidity", y="temperature")
```

```python
sns.jointplot(data=weather, x="humidity", y="temperature", kind="hex")
```

```python
sns.jointplot(data=weather, x="humidity", y="temperature", kind="kde")
```

```python
sns.pairplot(data=weather, vars=["humidity", "temperature", "air_pollution_index"])
```

```python
sns.stripplot(data=weather, x="weather_type", y="temperature")
```

```python
sns.lmplot(data=weather, x="humidity", y="temperature")
```

```python
sns.lmplot(data=weather, x="humidity", y="temperature", hue="weather_type")
```
